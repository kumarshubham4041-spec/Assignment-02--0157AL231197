```python
import os

# Global variables
users = {}  # Dictionary to store multiple users: {username: {'address':..., 'password':...}}
logged_user = None

# Function to clear console
def clear_screen():
    os.system('cls' if os.name == 'nt' else 'clear')

# Registration function
def register():
    clear_screen()
    print("\n--- Registration ---")
    while True:
        username = input("Enter your name: ").strip()
        if username in users:
            print("User already exists. Try a different name.")
        elif not username:
            print("Name cannot be empty.")
        else:
            break

    address = input("Enter your address: ").strip()
    
    while True:
        password = input("Set your password (min 6 chars): ").strip()
        if len(password) < 6:
            print("Password must be at least 6 characters long.")
        else:
            break

    users[username] = {'address': address, 'password': password}
    print(f"\nRegistration successful! Welcome, {username}.")

# Login function
def login():
    global logged_user
    clear_screen()
    print("\n--- Login ---")
    if logged_user:
        print(f"You are already logged in as {logged_user}. Logout first to switch user.")
        return

    username = input("Enter your name: ").strip()
    password = input("Enter password: ").strip()

    if username in users and users[username]['password'] == password:
        logged_user = username
        print(f"\nWelcome, {username}! You are now logged in.")
    else:
        print("\nInvalid credentials. Please try again.")

# Show profile function
def show_profile():
    clear_screen()
    if logged_user:
        print("\n--- Your Profile ---")
        print(f"Name: {logged_user}")
        print(f"Address: {users[logged_user]['address']}")
    else:
        print("\nPlease login first to view your profile.")

# Update profile function
def update_profile():
    global users
    clear_screen()
    if not logged_user:
        print("\nPlease login first to update profile.")
        return

    print("\n--- Update Profile ---")
    new_address = input("Enter new address (leave blank to keep current): ").strip()
    if new_address:
        users[logged_user]['address'] = new_address

    change_pass = input("Do you want to change your password? (y/n): ").lower()
    if change_pass == 'y':
        while True:
            new_password = input("Enter new password (min 6 chars): ").strip()
            if len(new_password) < 6:
                print("Password must be at least 6 characters long.")
            else:
                users[logged_user]['password'] = new_password
                print("Password updated successfully.")
                break

    print("Profile updated successfully!")

# Logout function
def logout():
    global logged_user
    clear_screen()
    if logged_user:
        print(f"{logged_user} has been logged out.")
        logged_user = None
    else:
        print("No user is currently logged in.")

# Exit function
def terminate():
    clear_screen()
    print("\nThank you for using the Student Management System. Goodbye!")
    exit()

# Main menu function
def main():
    while True:
        print("\n" + "="*50)
        print(" Welcome to LNCT Student Management System ")
        status = f"Logged in as: {logged_user}" if logged_user else "Status: Logged out"
        print(status)
        print("="*50)
        print("""
1. Registration
2. Login
3. Show Profile
4. Update Profile
5. Logout
6. Refresh Menu
7. Exit
""")
        choice = input("Select option (1-7): ").strip()

        if choice == '1':
            register()
        elif choice == '2':
            login()
        elif choice == '3':
            show_profile()
        elif choice == '4':
            update_profile()
        elif choice == '5':
            logout()
        elif choice == '6':
            clear_screen()
            continue
        elif choice == '7':
            terminate()
        else:
            print("Invalid option. Please select a number from 1 to 7.")

if __name__ == "__main__":
    main()
```
